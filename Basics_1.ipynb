# Operators

##### An operator in Python is a special symbol that performs an action on one or more values (variables). Operators are used for mathematical calculations, comparisons, logical operations, and more.

### Arithematic Operatoers

###### +  (Addition)       →  5 + 3  = 8  
###### *  (Multiplication) →  5 * 3  = 15  
###### /  (Division)       →  5 / 2  = 2.5  
###### %  (Modulus)        →  5 % 2  = 1  (Remainder)  
###### ** (Exponentiation) →  2 ** 3 = 8  (Power)  
###### // (Floor Division) →  5 // 2 = 2  (Removes decimal) 

### Comparison Operators (For comparing values, returns True or False)

##### ==  (Equal to)        →  5 == 5  → True  
##### !=  (Not equal to)    →  5 != 3  → True  
##### >   (Greater than)    →  5 > 3   → True  
##### <   (Less than)       →  5 < 3   → False  
##### >=  (Greater or equal) →  5 >= 5 → True  
##### <=  (Less or equal)   →  5 <= 4  → False 

### Logical Operators (For combining conditions)

###### and  → True if both conditions are true  → (5 > 2 and 10 > 5) → True  
###### or   → True if at least one condition is true  → (5 > 2 or 10 < 5) → True  
###### not  → Reverses the result  → not(5 > 2) → False  

### Assignment Operators (For assigning values)

###### =   (Assign)      →  x = 5  
###### +=  (Add & assign) →  x += 3  (x = x + 3)  
###### -=  (Subtract & assign) →  x -= 2  (x = x - 2)  
###### *=  (Multiply & assign) →  x *= 4  (x = x * 4)  


### Membership Operators (For checking values in sequences)

###### in     → Checks if value exists  →  "a" in "apple"  → True  
###### not in → Checks if value does not exist  →  5 not in [1,2,3] → True  

```python
fruits = ["apple", "banana", "cherry"]
print("apple" in fruits)  # Output: True
print("grape" in fruits)  # Output: False
```

```python
numbers = (1, 2, 3, 4, 5)
print(10 not in numbers)  # Output: True
print(3 not in numbers)   # Output: False
```

##### Checking Keys in a Dictionary

```python
person = {"name": "Naveen", "age": 25}
print("name" in person)      # Output: True (checks keys, not values)
print("Naveen" in person)    # Output: False
print("city" not in person)  # Output: True
```

## Identity Operators (For comparing memory locations of objects)

##### Identity operators are used to compare the memory locations of two objects.

###### is     → True if objects are the same  → x is y  
###### is not → True if objects are different  → x is not y  

```python
a = 10
b = 10
print(a is b)      # Output: True (Same memory location for small numbers)
print(a is not b)  # Output: False
```

```python
a = 1000
b = 1000
print(a is b)      
print(a is not b)
```

```python
list1 = [1, 2, 3]
list2 = [1, 2, 3]
print(list1 is list2)      # Output: False (Different memory locations)
print(list1 == list2)      # Output: True (Values are same)
print(list1 is not list2)  # Output: True
```
